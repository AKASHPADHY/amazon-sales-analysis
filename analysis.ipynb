```python
cleaned_data = sales_data.drop(columns=['index', 'New', 'PendingS'],errors='ignore')
cleaned_data
```

```python
missing_data=cleaned_data.isnull().sum()
missing_data
```

**handling missing value**

```python
cleaned_data = cleaned_data.dropna(subset=['Amount','currency'])
```

```python
cleaned_data[['ship-city', 'ship-state', 'ship-country']] = cleaned_data[['ship-city', 'ship-state', 'ship-country']].fillna("Unknown")
cleaned_data['ship-postal-code'] = cleaned_data['ship-postal-code'].fillna(-1)
cleaned_data = cleaned_data.drop(columns=['fulfilled-by'])
```

```python
cleaned_data.info(),cleaned_data.isnull().sum()
```

**Sales Overview**

```python
cleaned_data['Date'] = pd.to_datetime(cleaned_data['Date'], errors='coerce')
cleaned_data['Date']
```

```python
# Group sales data by month to observe trends
sales_trend = cleaned_data.groupby(cleaned_data['Date'].dt.to_period('M')).agg({'Amount': 'sum'}).reset_index()
sales_trend['Date'] = sales_trend['Date'].dt.to_timestamp()
sales_trend
```

```python
sales_trend.to_csv('sales_trend.csv')
```

```python
# Plot the sales trend over time
plt.figure(figsize=(12, 6))
plt.plot(sales_trend['Date'], sales_trend['Amount'], marker='o')
plt.title('Monthly Sales Trend', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales (Amount)', fontsize=12)
plt.grid()
plt.tight_layout()
plt.savefig('Monthly Sales Trend.jpg')
plt.show()
```

**product analysis**

```python
top_categories = cleaned_data.groupby('Category').agg({'Amount': 'sum', 'Qty': 'sum'}).sort_values(by='Amount', ascending=False).head(10)
top_categories
```

```python
top_categories.to_csv('top_categories.csv')
```

```python
#sales by categories
plt.figure(figsize=(12, 6))
sns.barplot(x=top_categories.index, y=top_categories['Amount'])
plt.title('Top Product Categories by Sales', fontsize=16)
plt.ylabel('Total Sales (Amount)', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.tight_layout()
plt.savefig('Total Sales (Amount).jpg')
plt.show()
```

```python
#sales by size
size_distribution = cleaned_data.groupby('Size').agg({'Qty': 'sum'}).sort_values(by='Qty', ascending=False)
size_distribution.to_csv('size_distribution.csv')

plt.figure(figsize=(10, 5))
sns.barplot(x=size_distribution.index, y=size_distribution['Qty'])
plt.title('Product Size Distribution', fontsize=16)
plt.ylabel('Total Quantity Sold', fontsize=12)
plt.xlabel('Size', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Product Size Distribution.jpg')
plt.show()
```

**geographical Analysis**

```python
# Sales by State
state_sales = cleaned_data.groupby('ship-state').agg({'Amount': 'sum', 'Qty': 'sum'}).sort_values(by='Amount', ascending=False)

plt.figure(figsize=(20, 6))
sns.barplot(x=state_sales.index, y=state_sales['Amount'])
plt.title('Sales by State', fontsize=16)
plt.ylabel('Total Sales (Amount)', fontsize=12)
plt.xlabel('State', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Sales by State.jpg')
plt.show()

# Top Cities by Sales
city_sales = cleaned_data.groupby('ship-city').agg({'Amount': 'sum'}).sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=city_sales.index, y=city_sales['Amount'])
plt.title('Top Cities by Sales', fontsize=16)
plt.ylabel('Total Sales (Amount)', fontsize=12)
plt.xlabel('City', fontsize=12)
plt.tight_layout()
plt.savefig('Top Cities by Sales.jpg')
plt.show()
```

**Customer Segmentation Approach**

```python
# Aggregate sales data by B2B flag
b2b_segmentation = cleaned_data.groupby('B2B').agg({
    'Amount': ['sum', 'mean'],
    'Qty': ['sum', 'mean'],
    'Order ID': 'count'
}).reset_index()

b2b_segmentation  
```

```python
b2b_segmentation.columns = ['B2B', 'Total Sales', 'Avg Sales Per Order', 'Total Quantity', 'Avg Quantity Per Order', 'Order Count']
b2b_segmentation
```

```python
b2b_segmentation.to_csv('b2b_segmentation.csv')
```

```python
# Aggregate sales by Sales Channel
channel_segmentation = cleaned_data.groupby('Sales Channel').agg({
    'Amount': 'sum',
    'Qty': 'sum',
    'Order ID': 'count'
}).sort_values(by='Amount', ascending=False).reset_index()
channel_segmentation.columns = ['Sales Channel', 'Total Sales', 'Total Quantity', 'Order Count']
channel_segmentation
```

```python
channel_segmentation.to_csv('channel_segmentation.csv')
```

```python
# Repeat transactions by state
repeat_transactions = cleaned_data.groupby(['ship-state', 'ship-city']).agg({
    'Order ID': 'count',
    'Amount': 'sum'
}).sort_values(by='Order ID', ascending=False).head(10)
repeat_transactions
```

```python
repeat_transactions.to_csv('repeat_transactions.csv')
```
